##### Closed form solution implementation

The cell performs standard linear regression on data using the closed form solution implementation.

```python
def standRegres(xArr,yArr):
    xMat = np.mat(xArr); yMat = np.mat(yArr).T
    xTx = xMat.T*xMat
    if np.linalg.det(xTx) == 0.0:
        print("This matrix is singular, cannot do inverse")
        return
    ws = xTx.I * (xMat.T*yMat)
    return ws
```

In order to perform the multiple regression, we need to add a column of 1s as the coefficient for x0. Passing the values to closed form model.

```python
# In order to do multiple regression we need to add a column of 1s as the coefficient for x0
np.set_printoptions(precision=6, linewidth=120, suppress=True, edgeitems=7)
X_train = np.array(X_train)
X_train = np.array([np.concatenate((v,[1])) for v in X_train])
y_train = np.array(y_train)
X_test = np.array(X_test)
X_test = np.array([np.concatenate((v,[1])) for v in X_test])
y_test = np.array(y_test)
```

```python
def cross_validate_closed_form(model, x, y, n, verbose=False):
    # model: regression model to be trained
    # X: the data matrix
    # y: the target variable array
    # n: the number of fold for x-validation
    # Returns mean RMSE across all folds
    
    from sklearn.model_selection import KFold
    kf = KFold(n_splits=n, random_state=22)
    xval_err = 0
    f = 1
    for train,test in kf.split(x):
        w = standRegres(x[train],y[train])
        #print(w)
        p = np.mat(x[test]) * w
        e = p.A.ravel()-y[test]
        rmse = np.sqrt(np.dot(e,e)/len(x[test]))
        if verbose:
            print("Fold %2d RMSE: %.4f" % (f, rmse))
        xval_err += rmse
        f += 1
    return xval_err/n
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet, SGDRegressor
from sklearn import feature_selection
from sklearn.model_selection import cross_val_score
from sklearn import metrics
```

Calculates the error associated with the training data.

```python
def measure_performance(X, y, clf):
    y_pred = clf.predict(X)
    MAE = metrics.mean_absolute_error(y, y_pred)
    return MAE
```

Returns the optimal features for the dataset and the weights associated with best features. Also plots the validation accuracy along with percentage of features.

```python
def best_feature(X, y, clf,vs_column, X_test, y_test, k):
    percentiles = range(1, 100, 5)
    results = []
    print("Average score for each percentile value\n")
    for i in range(1, 100, 5):
        fs = feature_selection.SelectPercentile(feature_selection.f_regression, percentile=i)
        X_train_fs = fs.fit_transform(X, y)
        scores = (cross_val_score(clf, X_train_fs, y, cv=k, scoring='neg_mean_absolute_error'))
        print("%2d  %0.4f" % (i, abs(scores.mean())))
        results = np.append(results, scores.mean())
    optimal_percentile_ind = np.where(results == results.max())[0][0]
    print("\nOptimal percentile of features:{0}".format(percentiles[optimal_percentile_ind]), "\n")
    optimal_num_features = int(percentiles[optimal_percentile_ind]*len(vs_column.columns)/100)
    print("Optimal number of features:{0}".format(optimal_num_features), "\n")
    pl.figure(figsize=(10,6))
    ax = plt.subplot()
    ax = plt.subplot(111)
    pl.xlabel("Percentage of features selected")
    pl.ylabel("Negative mean absolute error")
    pl.title("Percentage of features selected vs. Negative mean absolute error")
    pl.plot(percentiles,abs(results))
    fs_variables = feature_selection.SelectPercentile(feature_selection.f_regression, optimal_num_features)
    x_features = fs_variables.fit_transform(X, y)
    columns = np.asarray(vs_column.columns)
    support = np.asarray(fs_variables.get_support())
    columns_with_support = columns[support]
    print("Columns with most important features\n")
    print(columns_with_support,"\n")
    print("Columns with most important features along with weights\n")
    for i in range(len(vs_column.columns)):
        if fs_variables.get_support()[i]:
            print("%10s  %3.2f" % (vs_column.columns.values[i], fs_variables.scores_[i]))
    fs = feature_selection.SelectKBest(feature_selection.f_regression, optimal_num_features)
    Xtrain_fs = fs.fit_transform(X, y)
    clf.fit(Xtrain_fs, y)
    X_fs = fs.transform(X)
    mae = measure_performance(X_fs, y, clf)
    print('\nEvaluate performance with the new feature set on train data')
    print("Mean Absolute Error for train dataset: ", mae)
    X_train_fs = fs.fit_transform(X, y)
    clf.fit(X_train_fs, y)
    X_test_fs = fs.transform(X_test)
    mae = measure_performance(X_test_fs, y_test, clf)
    print('\nEvaluate performance with the new feature set on test data')
    print("Mean Absolute Error for test dataset: ", mae)
    
```

```python
from sklearn.model_selection import KFold

def calc_params(X, y, clf, param_values, param_name, K):
    
    # Convert input to Numpy arrays
    X = np.array(X)
    y = np.array(y)

    # initialize training and testing score arrays with zeros
    train_scores = np.zeros(len(param_values))
    test_scores = np.zeros(len(param_values))
    alpha = np.zeros(len(param_values))
    
    # iterate over the different parameter values
    for i, param_value in enumerate(param_values):
        print(param_name, ' = ', param_value)
        
        # set classifier parameters
        clf.set_params(**{param_name:param_value})
        
        # initialize the K scores obtained for each fold
        k_train_scores = np.zeros(K)
        k_test_scores = np.zeros(K)
        
        # create KFold cross validation
        cv = KFold(n_splits=K, shuffle=True, random_state=0)
        
        # iterate over the K folds
        j = 0
        for train, test in cv.split(X):
            # fit the classifier in the corresponding fold
            # and obtain the corresponding accuracy scores on train and test sets
            clf.fit(X[train], y[train])
            k_train_scores[j] = measure_performance(X[train], y[train], clf)
            k_test_scores[j] = measure_performance(X[test], y[test], clf)
            j += 1
            
        # store the mean of the K fold scores
        train_scores[i] = np.mean(k_train_scores)
        test_scores[i] = np.mean(k_test_scores)
        alpha[i]=float(param_value)
       
    # plot the training and testing scores in a log scale
    plt.figure(figsize=(10,6))
    plt.plot(param_values, train_scores, label='Train', alpha=0.4, lw=2, c='b')
    plt.plot(param_values, test_scores, label='X-Val', alpha=0.4, lw=2, c='g')
    plt.legend(loc=1)
    plt.xlabel(param_name + " values")
    plt.ylabel("Mean absolute error")
    pl.title("Mean absolute error vs. " + param_name + " values")

    # return the training and testing scores on each parameter value
    return alpha, train_scores, test_scores
```

###### Cross Validation 
Function to perform cross validation on dataset

```python
def cross_validate(model, X, y, n, verbose=False):
    # model: regression model to be trained
    # X: the data matrix
    # y: the target variable array
    # n: the number of fold for x-validation
    # Returns mean RMSE across all folds
    
    from sklearn.model_selection import KFold
    kf = KFold(n_splits=n, random_state=22)
    xval_err = 0
    f = 1
    for train,test in kf.split(X):
        model.fit(X[train],y[train])
        p = model.predict(X[test])
        e = p-y[test]
        rmse = np.sqrt(np.dot(e,e)/len(X[test]))
        if verbose:
            print("Fold %2d RMSE: %.4f" % (f, rmse))
        xval_err += rmse
        f += 1
    return xval_err/n
```

##### Evaluating model
Function to evaluate the RMSE value on the training set and testing set

```python
def evaluate_model(model, xtrain, ytrain, xtest, ytest):
    model.fit(xtrain,ytrain)
    p = model.predict(xtrain)
    e = p-ytrain
    total_train_error = np.dot(e,e)
    rmse_train = np.sqrt(total_train_error/len(p))
    
    ptest = model.predict(xtest)
    etest = ptest-ytest
    total_train_error = np.dot(etest,etest)
    rmse_test = np.sqrt(total_train_error/len(ptest))
    print('Method: %s' %model)
    print('RMSE on train set: %.4f' %rmse_train)
    print('RMSE on test set: %.4f' %rmse_test)
```
