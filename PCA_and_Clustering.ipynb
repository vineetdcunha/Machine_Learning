```python
def cluster_sizes(clusters,data):
    #clusters is an array of cluster labels for each instance in the data
    
    size = {}
    cluster_labels = np.unique(clusters)
    n_clusters = cluster_labels.shape[0]

    for c in cluster_labels:
        size[c] = len(data[clusters == c])
    return size
```

```python
size = cluster_sizes(clusters,vs_train_norm)

for c in size.keys():
    print("Size of Cluster", c, "= ", size[c])
```

Mean value for silhouettes

```python
def plot_silhouettes(data, clusters, metric='euclidean'):
    
    from matplotlib import cm
    from sklearn.metrics import silhouette_samples

    cluster_labels = np.unique(clusters)
    n_clusters = cluster_labels.shape[0]
    silhouette_vals = metrics.silhouette_samples(data, clusters, metric='euclidean')
    c_ax_lower, c_ax_upper = 0, 0
    cticks = []
    for i, k in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[clusters == k]
        c_silhouette_vals.sort()
        c_ax_upper += len(c_silhouette_vals)
        color = cm.jet(float(i) / n_clusters)
        pl.barh(range(c_ax_lower, c_ax_upper), c_silhouette_vals, height=1.0, 
                      edgecolor='none', color=color)

        cticks.append((c_ax_lower + c_ax_upper) / 2)
        c_ax_lower += len(c_silhouette_vals)
    
    silhouette_avg = np.mean(silhouette_vals)
    pl.axvline(silhouette_avg, color="red", linestyle="--") 

    pl.yticks(cticks, cluster_labels)
    pl.ylabel('Cluster')
    pl.xlabel('Silhouette coefficient')
    pl.title('Silhouette coefficient vs. Cluster')
    pl.tight_layout()
    #pl.savefig('images/11_04.png', dpi=300)
    pl.show()
    
    return
```

```python
plot_silhouettes(vs_train_norm, clusters)
```

The above cell shows the plot of various clusters vs. the Silhouette Coefficient. We see all the clusters are above the mean line without much interference.

In order to capture 95% variance, we would need at least 7 components.

```python
meanVals = np.mean(vs_train_norm, axis=0)
meanRemoved = vs_train_norm - meanVals #remove mean
covMat = np.cov(meanRemoved, rowvar=0)

np.set_printoptions(precision=2,suppress=True,linewidth=100)
print(covMat)
```

```python
import numpy.linalg as la
eigVals,eigVects = la.eig(np.mat(covMat))
```

```python
print(eigVals)
```

```python
print(eigVects)
```

Following are the eigenvalues and eigenvectors associated with each component.

```python
eigValInd = np.argsort(eigVals)  #sort, sort goes smallest to largest
eigValInd = eigValInd[::-1]   #reverse
sortedEigVals = eigVals[eigValInd]
print(sortedEigVals)
total = sum(sortedEigVals)
varPercentage = sortedEigVals/total*100
print(varPercentage)
```

In order to capture 95% variance, we would need at least 7 components.

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(range(1, 11), varPercentage[:10], marker='^')
plt.xlabel('Principal Component Number')
plt.ylabel('Percentage of Variance')
plt.title ('Percentage of Variance vs. Principal Component Number')
plt.show()
```

The following scree plot shows the curve flatening around 4 components. Since we need to capture 95% variance, we are consuming 7 components.
